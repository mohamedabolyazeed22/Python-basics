```python
# Problem 1: Managing a Shopping List
def shopping_list_manager():
    print("Welcome to the Shopping List Manager!")
    shopping_list = []
    
    while True:
        item = input("Add an item to your shopping list (or type 'done' to finish): ").strip()
        if item.lower() == 'done':
            break
        shopping_list.append(item)
        print(f"'{item}' added to your shopping list.")
    
    print("\nYour final shopping list:")
    for i, item in enumerate(shopping_list, 1):
        print(f"{i}. {item}")
    print("Happy shopping!\n")

# Problem 2: String Manipulation
def string_manipulator():
    print("\nWelcome to the String Manipulator!")
    my_string = "Hello world, welcome to Python programming!"
    print(f"Original string: {my_string}")
    
    # Uppercase
    upper_string = my_string.upper()
    print(f"Uppercase string: {upper_string}")
    
    # Replace
    replace_string = my_string.replace("world", "universe")
    print(f"Replaced string: {replace_string}")
    
    # Split
    split_string = my_string.split()
    print(f"List of words: {split_string}\n")

# Problem 3: Working with Sets
def set_operations():
    print("\nWelcome to the Set Operations!")
    my_set = {"apple", "banana", "orange"}
    print(f"Initial set: {my_set}")
    
    # Add a value
    add_value = input("Enter a new value to add to the set: ").strip()
    my_set.add(add_value)
    print(f"Set after adding '{add_value}': {my_set}")
    
    # Remove a value
    remove_value = input("Enter a value to remove from the set: ").strip()
    if remove_value in my_set:
        my_set.remove(remove_value)
        print(f"Set after removing '{remove_value}': {my_set}")
    else:
        print(f"'{remove_value}' not found in the set. No changes made.\n")

# Problem 4: Dictionary Operations
def student_grades_manager():
    print("\nWelcome to the Student Grades Manager!")
    students = {}
    
    # Add students
    for i in range(3):
        name = input(f"Enter student {i+1}'s name: ").strip()
        grade = input(f"Enter {name}'s grade: ").strip()
        students[name] = grade
        print(f"Added {name} with grade {grade}.")
    
    print("\nCurrent student grades:")
    for name, grade in students.items():
        print(f"{name}: {grade}")
    
    # Search for a student
    search_name = input("\nEnter a student's name to search for their grade: ").strip()
    if search_name in students:
        print(f"{search_name}'s grade is {students[search_name]}.")
    else:
        print(f"'{search_name}' not found in the records.")
    
    # Update a student's grade
    update_name = input("\nEnter a student's name to update their grade: ").strip()
    if update_name in students:
        new_grade = input(f"Enter the new grade for {update_name}: ").strip()
        students[update_name] = new_grade
        print(f"Updated {update_name}'s grade to {new_grade}.")
    else:
        print(f"'{update_name}' not found in the records. No changes made.")
    
    print("\nFinal student grades:")
    for name, grade in students.items():
        print(f"{name}: {grade}")
    print("Thank you for using the Student Grades Manager!\n")

# Main Program
def main():
    print("Welcome to the Python Practice Program!")
    while True:
        print("\nChoose a problem to solve:")
        print("1. Shopping List Manager")
        print("2. String Manipulator")
        print("3. Set Operations")
        print("4. Student Grades Manager")
        print("5. Exit")
        
        choice = input("Enter your choice (1-5): ").strip()
        
        if choice == '1':
            shopping_list_manager()
        elif choice == '2':
            string_manipulator()
        elif choice == '3':
            set_operations()
        elif choice == '4':
            student_grades_manager()
        elif choice == '5':
            print("Thank you for using the Python. Goodbye!")
            break
        else:
            print("Invalid choice. Please try again.")

if __name__ == "__main__":
    main()
```
